

Replacing the Null values with "Unknown":

```python
# Dropping null values as these are very less in number

df.dropna(subset=['date_added', 'rating'], inplace = True)
```

```python
df['date_added_clean'] = df['date_added'].str.strip()
```

```python
df['date_added_datetime'] = pd.to_datetime(df['date_added_clean'], format="%B %d, %Y")
```
